## <span style="color:#2774AE;">🧠 The Basic Model of a Computer</span>

To understand programming, it helps to view a computer as a system composed of two primary functional parts working together: **Memory** and the **Logical Unit**.

---

### **1. The Two-Part Architecture**

* **Memory (RAM):** Think of this as a vast collection of numbered "storage bins." This is where the computer keeps the data it needs to work with.
* **The Logical Unit (CPU):** This is the "brain" or the engine. It performs calculations, makes comparisons, and follows instructions. It cannot "remember" much on its own; it must pull data from memory to process it.



---

### **2. The Process of Programming**

When you write and run a Python script, you are essentially managing the flow between these two parts in a two-step cycle:

#### **Step A: Initializing Data (Storage)**
First, we put data into the **Memory**. In Python, we do this using **Variables**.
> *Example:* `x = 10`  
> This tells the computer: *"Find a spot in memory, label it 'x', and store the number 10 there."*

#### **Step B: Performing Operations (Processing)**
Next, we tell the **Logical Unit** to do something with that stored data.
> *Example:* `y = x + 5`  
> This tells the computer: *"Go to memory, find the value labeled 'x', bring it to the Logical Unit, add 5 to it, and store the new result in a memory spot labeled 'y'."*

---

### **🔄 Summary: The Workflow**
1. **Input:** Data is stored in Memory.
2. **Process:** The Logical Unit manipulates that data based on your code.
3. **Output:** The result is either stored back in Memory or sent to your screen (like our `print` command!).

## <span style="color:#2774AE;">💾 Data Initialization and Types</span>

In Python, **Initialization** is the process of assigning a value to a variable and storing it in the computer's memory. When you initialize data, Python automatically decides how much memory to allocate based on the **Data Type**.

Different types require different amounts of "storage space." For example, a simple true/false value (`bool`) needs much less memory than a long list of coordinates or a high-resolution text string.

---

### **Core Data Types in Python**

Throughout this workshop, we will interact with these primary categories:

* **Text Type:** `str` (Strings - used for sequences of characters)
* **Numeric Types:** * `int`: Whole numbers (integers)
    * `float`: Decimal numbers (floating-point)
    * `complex`: Complex numbers (used in advanced biophysics/math)
* **Sequence Types:** * `list`: Ordered, changeable collections
    * `tuple`: Ordered, unchangeable collections
    * `range`: Often used for loops
* **Mapping Type:** `dict` (Dictionaries - key-value pairs like a real-world dictionary)
* **Boolean Type:** `bool` (Logical values: `True` or `False`)



---

> **💡 Note on Memory:** Python is "dynamically typed," meaning you don't have to tell it how much memory to use. It looks at the value you provide (e.g., `3.14`) and automatically initializes a `float` in memory for you.


````python
import sys

# 1. Initialize the data in memory
my_bool = True
my_int = 1000
my_float = 3.14159
my_string = "UCLA QCBio"
my_emptylist = []
my_list = [1, 2, 3]
my_boollist = [True, False, True]
my_floatlist = [1.1, 2.2, 3.3]
my_longlist = [1,2,3,4,5,6,7,8]

# 2. Query the memory size (in Bytes)
print(f"Memory for Boolean:  {sys.getsizeof(my_bool)} bytes")
print(f"Memory for Integer:  {sys.getsizeof(my_int)} bytes")
print(f"Memory for Float:    {sys.getsizeof(my_float)} bytes")
print(f"Memory for String:   {sys.getsizeof(my_string)} bytes")
print(f"Memory for EmptyList:{sys.getsizeof(my_emptylist)} bytes")
print(f"Memory for List:     {sys.getsizeof(my_list)} bytes")
print(f"Memory for BoolList: {sys.getsizeof(my_boollist)} bytes")
print(f"Memory for FloatList:{sys.getsizeof(my_floatlist)} bytes")
print(f"Memory for LongList: {sys.getsizeof(my_longlist)} bytes")
````

## <span style="color:#2774AE;">🌍 Initialization Across Languages</span>

As we’ve seen, **Initialization** is just the act of telling the computer: *"Reserve a spot in memory and put this value inside it."* While the logic is identical, different languages have different "syntax" (rules) for how to write that request.

---

### **Comparing the "Storage Command"**

In Python, we don't have to tell the computer what type of data we are using; it figures it out automatically. In many other languages used in science (like C++ or Java), you must be much more explicit.

| Language | Code for Initialization | The Computer's Interpretation |
| :--- | :--- | :--- |
| **Python** | `x = 5` | "Find a bin, name it `x`, and store an integer `5`." |
| **JavaScript**| `let x = 5;` | "Declare a variable `x` and assign it the number `5`." |
| **C++** | `int x = 5;` | "Reserve exactly 4 bytes of integer space and put `5` there." |
| **R** | `x <- 5` | "Assign the value `5` to the object `x`." |



---

### **Static vs. Dynamic Initialization**

There are two main ways languages handle their "Memory Bins":

1. **Dynamic (Python, R, JavaScript):** You can change the type of data in a bin whenever you want. 
   *Example:* You can say `x = 5` and then later say `x = "UCLA"`. The computer simply replaces the integer with a string.

2. **Static (C, C++, Java):** Once you label a bin as an `int`, it can **only** hold integers.
   *Example:* If you try to put the text "UCLA" into an `int x` bin, the computer will throw a "Type Error" before the program even runs.

---

### **Research Tip**
If you move from Python to a language like C++ or Java, remember that the computer becomes even **more** literal. If you forget to state the data type (like `int` or `float`) during initialization, the computer will refuse to even look at the rest of your code!

> **Why Python is popular in Science:** Because Python handles the "type" and "memory size" for you, researchers can focus on the logic of their experiments rather than the plumbing of the computer's memory.



## <span style="color:#2774AE;">📝 Documentation: Using Comments</span>

In programming, **Comments** are notes written within the code that are intended for humans to read. The Python interpreter completely ignores these lines.

---

### **1. Single-Line Comments**
Any text following a **hash sign (`#`)** is treated as a comment. This is the most common way to provide a brief explanation of a specific line of code.

```python

```

### **2. Multiline Comments (Docstrings)**

For longer explanations that span multiple lines, you can use **triple quotes** (`"""` or `'''`).

Technically, this is a "string literal." However, if a string is not assigned to a variable (meaning it is not stored in a memory bin), the Python interpreter will simply bypass it.

```python
'''
This is a multiline comment.
It is useful for describing the overall purpose 
of a script or a complex block of code.
'''
print("Hello, world!")

```

---

### **💡 Best Practices**

* **Explain the "Why":** Good comments don't just repeat what the code says; they explain the logic or the intent behind it.
* **Keep it Clean:** Use comments to keep your "Mental Model" organized so that when you revisit your code in the future, you understand your own thought process.

## <span style="color:#2774AE;">🔤 String Operations</span>

In Python, text is stored as a **String** (`str`). While strings are not numbers, we can still use certain operators to manipulate them, and we use special formatting to inject data directly into our text. String literals are surrounded by either a pair of single quotation marks (`'`) or a pair of double quotation marks (`"`).

---

### **1. Basic String "Arithmetic"**
Python allows you to use `+` and `*` with strings, but they perform **concatenation** and **repetition** rather than math.

| Operation | Syntax | Result |
| :--- | :--- | :--- |
| **Concatenation** | `"UCLA" + " " + "QCBio"` | `"UCLA QCBio"` |
| **Repetition** | `"DNA" * 3` | `"DNADNADNA"` |

---

### **2. The Power of f-Strings (Formatted Strings)**
The most modern and efficient way to handle strings in Python is the **f-string**. By placing an `f` before the opening quotation mark, you tell Python: *"Look inside this string for curly braces `{}` and replace them with actual variable values."*



#### **Why use f-strings?**
* **Readability:** You can see exactly what the final sentence will look like.
* **Speed:** f-strings are faster for the computer to process than older methods.
* **Auto-Conversion:** Python automatically converts numbers into text for you inside the braces.

#### **Example Code:**
```python
name = "Amelia"
score = 95.5

# Using an f-string to combine text and variables
message = f"Student {name} scored {score}% on the lab."

print(message)

```

---

### **3. Advanced f-String Tricks**

You can even perform math or call functions directly inside the curly braces!

```python
radius = 5
print(f"The area of the circle is {3.14159 * radius**2}")

sequence = "atgc"
print(f"The sequence in uppercase is: {sequence.upper()}")

```

---

### **🕵️ Warning: The Missing 'f'**

If you forget to put the `f` at the very beginning of the string, the computer becomes literal. It will print the curly braces and the variable name exactly as you typed them, rather than the data inside the memory bin.

```python
# Incorrect: Forgot the 'f'
x = 10
print("The value is {x}") # Output: The value is {x}

# Correct
print(f"The value is {x}") # Output: The value is 10

```

---

> **🔬 Research Tip:** When generating reports or naming files for 1,000 different samples, f-strings are your best friend. For example: `f"sample_{sample_id}_temp_{temperature}.csv"` allows you to automate file naming instantly.

## <span style="color:#2774AE;">✂️ String Manipulations: Cleaning and Combining</span>

Beyond basic addition and multiplication, Python provides built-in "methods" to manipulate strings. These are essential for cleaning up messy data (like removing extra spaces) or reformatting sequences.

---

### **1. `.strip()` — Removing Extra Space**
When reading data from a file, strings often come with invisible "whitespace" (spaces or new lines) at the beginning or end. `.strip()` acts like a pair of scissors, trimming off any leading or trailing whitespace.

```python
raw_data = "   ATGCATGC   "
clean_data = raw_data.strip()

print(f"Original: '{raw_data}'")
print(f"Cleaned:  '{clean_data}'")

```


### **2. `.count()` — Finding Patterns**

The `.count()` method allows the "Logical Unit" to scan a string and tell you how many times a specific character or pattern appears.

```python
sequence = "GATTACA"
count_A = sequence.count("A")

print(f"There are {count_A} 'A' bases in the sequence {sequence}.")
```


### **3. `.join()` — The "Glue" for Lists**

The `.join()` method is a bit unique. It takes a **list** of strings and glues them together into one single string using a specific separator.

```python
nucleotides = ["A", "T", "G", "C"]

# Use a hyphen as glue
hyphenated = "-".join(nucleotides)

# Use nothing as glue (empty string)
combined = "".join(nucleotides)

print(f"Hyphenated: {hyphenated}")
print(f"Combined:   {combined}")
```


### **📊 Summary Table**

| Method | Purpose | Example | Result |
| --- | --- | --- | --- |
| **`.strip()`** | Removes whitespace from ends | `" hello ".strip()` | `"hello"` |
| **`.count()`** | Counts occurrences | `"apple".count("p")` | `2` |
| **`.join()`** | Glues a list into a string | `",".join(["A", "B"])` | `"A,B"` |

---

### **🕵️ Strings are Immutable**

Remember: Strings are "Unchangeable" (Immutable). When you use `.strip()` or `.join()`, Python does **not** change the original variable. Instead, it creates a **brand new** string in a new memory bin.

```python
name = " UCLA "
name.strip() # This does nothing to the 'name' variable!
print(f"'{name}'") # Still has spaces

# You must re-assign it to save the change:
name = name.strip()
print(f"'{name}'") # Now it is clean!

```

## <span style="color:#2774AE;">⚖️ Boolean Logic: Asking the Computer Questions</span>

To make decisions in an `if` statement, we need to ask the computer questions that result in a **Boolean** (`True` or `False`). We do this using **Relational Operators** and **Logical Operators**.

---

### **1. Relational Operators**

These operators compare two values. Think of them as the "tests" the Logical Unit runs on your memory bins.

| Operator | Meaning | Example | Result |
| :--- | :--- | :--- | :--- |
| `==` | **Equal to** | `10 == 10` | `True` |
| `!=` | **Not equal to** | `10 != 5` | `True` |
| `>` | **Greater than** | `12 > 12` | `False` |
| `<` | **Less than** | `10 < 15` | `True` |
| `>=` | **Greater than or equal to** | `12 >= 12` | `True` |
| `<=` | **Less than or equal to** | `5 <= 2` | `False` |



#### **⚠️ Warning: `=` vs `==`**
One of the most common errors in programming is confusing `=` with `==`.
* `x = 10`: **Assignment.** "Take the number 10 and put it in a bin labeled `x`."
* `x == 10`: **Comparison.** "Look inside bin `x`. Is there a 10 in there? Answer Yes or No."

---

### **2. Logical Operators (AND, OR, NOT)**

Sometimes one question isn't enough. We use logical operators to combine multiple comparisons into a single decision.

* **`and`**: Returns `True` only if **both** statements are true.
* **`or`**: Returns `True` if **at least one** statement is true.
* **`not`**: Reverses the result (True becomes False).

| Operation | Logical Test | Result |
| :--- | :--- | :--- |
| **AND** | `(5 > 2) and (10 < 15)` | `True` (Both are true) |
| **OR** | `(5 < 2) or (10 < 15)` | `True` (One is true) |
| **NOT** | `not (5 == 5)` | `False` (Reverses True) |



---

### **3. Comparing Strings**
Python can also compare text. It does this using **alphabetical order** (technically ASCII/Unicode order).

```python
# Does "ab" come after "bb" in the dictionary?
print("ab" >= "bb")  # Result: False

# Are these sequences the same?
print("ATGC" == "atgc") # Result: False (Python is case-sensitive!)

```

## <span style="color:#2774AE;">🐜 Activity 6: The Ant Intersection Mystery</span>

Imagine two ants crawling along a thin wire (a 1D number line). We know their positions at two points in time. If they are traveling at constant speeds, we can determine if they will ever occupy the exact same spot at the same time.

### **The Goal**
Use the positions of Ant A and Ant B to calculate their speeds, then use an **if statement** to determine if they meet at a specific time.

---

## <span style="color:#2774AE;">🛣️ Control Flow: The `if` Statement</span>

An **if statement** allows your program to make decisions. By using Boolean logic (True/False), the "Logical Unit" can decide which "path" of code to execute and which to skip.

---

### **1. The Anatomy of a Decision**

In Python, we use `if`, `elif` (else if), and `else` to handle different scenarios.

```python
if <logical_statement>:
    # Executed if logical_statement is True
    print("Condition 1 is met.")

elif <logical_statement_2>: # (Optional)
    # Executed if statement 1 is False AND statement 2 is True
    print("Condition 2 is met.")

else: # (Optional)
    # Executed if all previous statements are False
    print("No conditions were met.")

```

---

### **2. The Golden Rule: Indentation**

Unlike many other languages (like C++ or Java) that use curly braces `{ }` or `end` keywords to group code, Python uses **indentation**.

* The colon `:` tells Python: "A new block of code is starting."
* The **indentation** (the empty space at the start of the line) tells Python which lines belong to that block.

#### **⚠️ Warning: Indentation Errors**

If your indentation is inconsistent, the computer will get confused and stop the program.

* **Standard Practice:** Always use **four spaces** for one level of indentation.
* Most modern editors (like VS Code or Jupyter) will automatically insert four spaces when you hit the `Tab` key.


---

### **3. Practical Example: DNA Analysis**

```python
gc_content = 55.0

if gc_content > 60:
    print("High GC-content detected.")
elif gc_content < 40:
    print("Low GC-content detected.")
else:
    print("Moderate GC-content.")

```

---

> **💡 Best Practice:** Always keep your indentation consistent. Mixing tabs and spaces is a common source of bugs that are very difficult to see with the naked eye!

## <span style="color:#2774AE;">⌨️ The `input()` Function: Making Code Interactive</span>

In a lab setting, you don't always want to hard-code your values. Sometimes you need to ask the researcher for the **Sample ID**, the **Incubation Temperature**, or the **Dilution Factor**.

The `input()` function allows your program to pause and wait for the user to type something.

---

### **1. How it Works**

The `input()` function always treats what the user types as a **String** (text), even if they type a number.

```python
# The text inside the parentheses is the 'prompt'
name = input("Enter the researcher's name: ")
print(f"Logged in as: {name}")

```

---

### **2. The "Number Trap" (Type Casting)**

Since `input()` always returns a string, you cannot do math with it immediately. You must convert it using `int()` (for whole numbers) or `float()` (for decimals).

```python
# ❌ This will cause an error if you try to do math
temp_str = input("Enter current incubator temp: ") 

# ✅ Convert to float for mathematical calculations
temp_float = float(temp_str)
target_temp = 37.0

difference = target_temp - temp_float
print(f"The incubator is {difference} degrees away from the target.")

```

---
