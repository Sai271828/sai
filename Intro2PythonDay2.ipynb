## <span style="color:#2774AE;">♻️ The `for` Loop: Automating Repetitive Tasks</span>

In programming, we rarely want to perform a task just once. Whether you are analyzing 10,000 cells or processing 50 blood samples, you need a way to tell the computer: *"Do this same action for every item in this list."*

---

### **1. Anatomy of a `for` Loop**

The `for` loop acts like a conveyor belt. It picks up one item at a time from a list 0r list-like object, stores it in a temporary variable, and executes a block of code.

```python
# General Syntax
for item in list:
    # Do something with 'item'
    print(item)

```

---

## <span style="color:#2774AE;">🦈 Activity 2: The 100-Step Predator-Prey Simulation</span>

In this activity, we will simulate a "pursuit" over 100 discrete time steps. We will track the distance at every second to see if the shark eventually catches the tuna or if the tuna escapes.

### **The Simulation Logic**

* **The Loop:** We will use `range(1, 101)` to represent 100 seconds of time.
* **The Update:** At every step, we recalculate the distance and update positions.
* **The "Catch" Condition:** We will add a `break` statement. If the distance becomes less than **1.0 unit**, the shark has caught the tuna!

---

### **Task: Complete the Simulation Script**

```python
# 1. Initial Positions
shark_x, shark_y = 0.0, 0.0
tuna_x, tuna_y = 30.0, 40.0 # Started a bit further away

# 2. Constant Velocities
s_vx, s_vy = 0.3 , 0.4 # When chasing the total speed is doubled
t_vx, t_vy = 0, 0.8  # Tuna is faster normally

detection_threshold = 25.0

print("🌊 SIMULATION STARTING...")

# 3. The Time Loop
for second in range(1, 101):
    # Calculate distance
    dist = ((tuna_x - shark_x)**2 + (tuna_y - shark_y)**2)**0.5
    
    # Check for "The Catch"
    if dist < 1.5:
        print(f"🍴 SUCCESS! The Shark caught the Tuna at second {second}!")
        break
    
    # Agent Logic: Chase vs. Cruise
    if dist < detection_threshold:
        # CHASE MODE: Shark orients itself towards Tuna and Tuna away from Shark
        shark_x += ...
        shark_y += ...
        tuna_x += ...
        tuna_y += ...
    else:
        # CRUISE MODE: Shark swims at constant velocity
        shark_x += ...
        shark_y += ...
        tuna_x += ...
        tuna_y += ...
    
    

    # Optional: Print status every 10 seconds to keep the console clean
    if second % 10 == 0:
        print(f"Step {second}: Distance is {dist:.2f} units.")

    # This runs ONLY if the loop finishes 100 steps without a 'break'

```

---



````python
shark_x, shark_y = 0.0, 0.0
tuna_x, tuna_y = 30.0, 40.0 # Started a bit further away

# 2. Constant Velocities
s_vx, s_vy = 0.3 , 0.4 # When chasing the total speed is doubled
t_vx, t_vy = 0, -0.8  # Tuna is faster normally

detection_threshold = 25.0

print("🌊 SIMULATION STARTING...")

# 3. The Time Loop
for second in range(1, 151):
    # Calculate distance
    dist = ((tuna_x - shark_x)**2 + (tuna_y - shark_y)**2)**0.5
    
    # Check for "The Catch"
    if dist < 1.5:
        print(f"🍴 SUCCESS! The Shark caught the Tuna at second {second}!")
        break
    
    # Agent Logic: Chase vs. Cruise
    if dist < detection_threshold:
        # CHASE MODE: Shark orients itself towards Tuna and Tuna away from Shark
        shark_x += (tuna_x - shark_x)/dist
        shark_y += (tuna_y - shark_y)/dist
        tuna_x += (tuna_x - shark_x)/dist*0.8
        tuna_y += (tuna_y - shark_y)/dist*0.8
    else:
        # CRUISE MODE: Shark swims at constant velocity
        shark_x += s_vx
        shark_y += s_vy
        tuna_x += t_vx
        tuna_y += t_vy
    
    

    # Optional: Print status every 10 seconds to keep the console clean
    if second % 10 == 0:
        print(f"Step {second}: Distance is {dist:.2f} units.")
        if dist < detection_threshold:
            print("  🦈 Shark is CHASING the Tuna!")

else:
    # This runs ONLY if the loop finishes 100 steps without a 'break'
    print("🐟 ESCAPED! The Tuna outswam the Shark for 100 seconds.")
````

## <span style="color:#2774AE;">🏎️ List Comprehension: Loops in Fast Forward</span>

Often in data science, we want to take one list, perform a small operation on every item, and save the results into a **new list**. While a standard `for` loop works perfectly for this, Python offers a "shortcut" called **List Comprehension**.

---

### **1. The Syntax Transformation**

Think of List Comprehension as a `for` loop compressed into a single line inside square brackets.

**The Standard Way:**

```python
new_list = []
for item in old_list:
    new_list.append(item * 2)

```

**The List Comprehension Way:**

```python
new_list = [item * 2 for item in old_list]

```

---

### **2. Why use it?**

* **Readability:** Once you are used to the syntax, it is much easier to read "at a glance."
* **Performance:** It is actually slightly faster for the computer to execute than a standard `append()` loop.
* **Cleanliness:** It prevents your script from being cluttered with dozens of empty lists (`list = []`) initialized before loops.

---

### **🧬 Processing Lab Measurements**

In this activity, imagine you have a list of temperatures recorded in **Celsius**, and you need to convert all of them to **Fahrenheit** for a publication.

#### **Task: Complete the Code**

```python
celsius_readings = [20.0, 25.5, 37.0, 40.2, 18.9]

# Use list comprehension to convert the whole list at once!
# Formula: (C * 9/5) + 32
fahrenheit_readings = [(c * 9/5) + 32 for c in celsius_readings]

print(f"Original (C): {celsius_readings}")
print(f"Converted (F): {fahrenheit_readings}")

```

---

### **🧪 Filtering with List Comprehension**

You can even add an `if` statement at the very end to "filter" your data.

**Example: Find only the "Fever" temperatures (above 37.5°C)**

```python
readings = [36.5, 37.1, 38.5, 39.2, 36.8]

fevers = [temp for temp in readings if temp > 37.5]

print(f"High temperatures detected: {fevers}")

```

## <span style="color:#2774AE;">🪆 Nested Loops: Loops Within Loops</span>

Sometimes, a single "conveyor belt" isn't enough. When you have data organized in a grid (like a spreadsheet) or you need to compare every item in one list to every item in another, you use a **Nested Loop**.

Think of this like a **clock**: The "inner loop" (the second hand) must complete a full circle of 60 seconds before the "outer loop" (the minute hand) moves just one notch.

---

### **1. The Mechanics of Nesting**

In a nested loop, for every **single** iteration of the outer loop, the inner loop runs to **completion**.

```python
for outer_item in outer_list:
    # This code runs once per outer_item
    for inner_item in inner_list:
        # This code runs MANY times
        print(f"Outer: {outer_item}, Inner: {inner_item}")

```


---

### **🧬 Activity 4: The Universal Donor Matchmaker**

* For this exercise, we will simplify: **Type O** can give to anyone.
* Anyone else can only give to a **Perfect Match**.

#### **Task: Complete the Code**

```python

print("--- Compatibility Report ---")

# Outer loop: Pick one donor bag at a time
for d_type in donors:
    
    # Inner loop: Check that bag against every single patient
    for p_type in patients:

        if d_type == "O" or d_type == p_type:
```

## <span style="color:#2774AE;">🪆 Nested List Comprehension</span>

A nested list comprehension follows the same order as a nested `for` loop. The **outer loop** comes first, followed by the **inner loop**.

**The Transformation:**

```python
# Standard Nested Loop
pairs = []
for x in ["A", "B"]:
    for y in [1, 2]:
        pairs.append((x, y))

# Nested List Comprehension
pairs = [(x, y) for x in ["A", "B"] for y in [1, 2]]

```

In Python, a **List** is an ordered collection of items stored in a single variable. Think of it like a rack of test tubes: each tube has a specific position (index), and you can add, remove, or swap tubes as the experiment progresses.

## <span style="color:#2774AE;">🧬 Sets: The Automated Duplicate Filter</span>

A **Set** is an unordered collection of **unique** items. In Python, sets are written with curly braces `{}`, just like dictionaries, but they only contain single values instead of key-value pairs.

In biology, sets are the ultimate tool for handling "Redundancy." If you have a list of 1,000 gene sequences but only want to know which *unique* species they came from, a Set is your best friend.

---

### **1. The Two Defining Features of Sets**

1. **Uniqueness:** A set cannot have the same item twice. If you try to add a duplicate, Python simply ignores it.
2. **Unordered:** Unlike a list, there is no "first" or "last" item. You cannot access items using an index like `my_set[0]`.

```python
# Creating a set with duplicates
observed_species = {"Lion", "Zebra", "Lion", "Elephant", "Zebra"}

print(observed_species)
# Output: {'Lion', 'Elephant', 'Zebra'}  <-- Duplicates are gone!

```

---

### **2. The Main Advantage: Membership Testing & Speed**

The "Superpower" of a set is its **speed**. Because of the way sets are stored in memory (using a technique called "Hashing"):

* **Lightning-Fast Lookups:** Checking if an item exists in a set (`if "DNA" in my_set`) takes the same amount of time whether the set has 10 items or 10,000,000 items.
* **Mathematical Operations:** You can instantly perform "Venn Diagram" math like **Unions**, **Intersections**, and **Differences**.

| Operation | Command | Biological Use Case |
| --- | --- | --- |
| **Union** | `set1 | set2` |
| **Intersection** | `set1 & set2` | Find genes that exist in **both** Treatment A and B. |
| **Difference** | `set1 - set2` | Find mutations present in the tumor but **not** in healthy tissue. |

---

### **🧬 Activity 6: Comparing Drug Resistance**

You are comparing two strains of bacteria to see which antibiotic resistance genes they share and which ones are unique.

#### **Task: Use Set Math**

```python
strain_a = {"Penicillin", "Tetracycline", "Erythromycin"}
strain_b = {"Tetracycline", "Vancomycin", "Methicillin"}

# 1. Which resistances do BOTH strains share? (Intersection)
shared = strain_a & strain_b
print(f"Common Resistance: {shared}")

# 2. Which resistances are ONLY in Strain A? (Difference)
unique_to_a = strain_a - strain_b
print(f"Unique to Strain A: {unique_to_a}")

# 3. Create a master list of all resistances found. (Union)
master_list = strain_a | strain_b
print(f"Master Inventory: {master_list}")

```

---

### **🕵️ Warning: The Empty Dictionary Trap**

If you want to create an empty set, you might be tempted to write `my_set = {}`.

**Don't!** Python will think you are creating an empty **Dictionary**. To create an empty set, you must use the `set()` function:

```python
# Wrong (This is a dictionary)
empty_thing = {} 

# Right (This is a set)
real_empty_set = set()

```

---

> **🔬 Research Tip:** If you have a massive CSV file with millions of rows and you need to remove duplicates, the fastest way in Python is to turn that column into a list and then wrap it in `set()`.
> `unique_data = list(set(massive_list))` — This is a common "trick" used in data cleaning.

## <span style="color:#2774AE;">🔢 The `range()` Function: Generating Sequences</span>

In Python, `range()` is the "Automated Counter" of the language. Instead of manually typing out a list of numbers like `[0, 1, 2, 3, 4]`, you use `range()` to generate them on the fly. This is most commonly used to tell a **for loop** exactly how many times to run.

---

### **1. How `range()` Works**

The `range()` function can take up to three arguments: **Start**, **Stop**, and **Step**.

| Syntax | Result | Example |
| --- | --- | --- |
| `range(stop)` | Counts from 0 up to (but **not** including) stop. | `range(5)`  0, 1, 2, 3, 4 |
| `range(start, stop)` | Counts from start up to (but **not** including) stop. | `range(2, 6)`  2, 3, 4, 5 |
| `range(start, stop, step)` | Counts by a specific increment (step). | `range(0, 10, 2)`  0, 2, 4, 6, 8 |

---

### **2. Using `range()` in Loops**

`range()` is perfect for when you need to perform an action a specific number of times, or when you need to use the **index** of a list.

```python
# Loop 3 times for 3 replicates
for i in range(1, 4):
    print(f"Running Replicate #{i}...")

# Output:
# Running Replicate #1...
# Running Replicate #2...
# Running Replicate #3...

```


### **🕵️ Warning: The "Off-By-One" Error**

One of the most confusing things for beginners is that `range()` is **exclusive** of the stop value.

* If you write `range(10)`, the last number produced is **9**.
* If you want to include **10**, you must write `range(11)`.

Think of it like a fence: the "stop" value is the last post of the fence, but you only care about the sections *before* it.

---

### **🔬 Advanced: `range()` vs `len()**`

When you want to loop through a list and change the items based on their position, you combine these two functions:

```python
concentrations = [1.0, 0.5, 0.25]

# range(len(list)) gives us the indices: 0, 1, 2
for i in range(len(concentrations)):
    # We can now modify the list at that specific index
    concentrations[i] = concentrations[i] * 10 

print(f"Updated Concentrations: {concentrations}")

```

---

> **🔬 Research Tip:** `range()` is incredibly memory efficient. It doesn't actually create a giant list of numbers in your computer's RAM; it just "remembers" where it started and what the next number should be. This allows you to write `range(1000000000)` without crashing your laptop!


## <span style="color:#2774AE;">🔄 The `while` Loop: Logic Until a Goal is Met</span>

While a `for` loop runs for a **fixed number of times**, a `while` loop runs **as long as a specific condition is True**.

In the lab, this is like an automated sensor: *"Keep stirring the solution **while** the temperature is below 50°C."*

---

### **1. Anatomy of a `while` Loop**

A `while` loop checks its condition *before* every iteration. If the condition becomes `False`, the loop stops immediately.

```python
while condition_is_true:
    # Do these actions
    # IMPORTANT: Change something so the condition eventually becomes False!

```

---

### **2. The Three Parts of a Safe `while` Loop**

To avoid a program that runs forever (an "Infinite Loop"), you need three things:

1. **Initial State:** A variable to check (e.g., `count = 0`).
2. **The Condition:** The rule for staying in the loop (e.g., `while count < 5`).
3. **The Update:** A way to change the variable inside the loop (e.g., `count += 1`).

---

### **🧬 Activity 8: Bacterial Growth Simulation**

Let's simulate a bacterial culture that doubles every hour. We want to know how many hours it takes for the population to exceed a "Saturation Point" of 1,000 cells.

#### **Task: Run the Simulation**

```python
population = 10      # Starting number of cells
hours = 0            # Starting time
saturation_limit = 1000

print(f"Starting Simulation... Initial Population: {population}")

# Keep looping as long as we are under the limit
while population < saturation_limit:
    population = population * 2  # Double the cells
    hours = hours + 1             # Increment time
    print(f"Hour {hours}: Population is now {population}")

print("-" * 30)
print(f"Goal Reached! It took {hours} hours to hit saturation.")

```

---

### **🕵️ Warning: The Infinite Loop**

If you forget the **Update** step (like `hours = hours + 1`), the condition will *always* be True. The computer will keep doubling the population forever, eventually crashing your program or freezing your computer.

* **Pro-Tip:** If your code gets stuck in an infinite loop, click in the terminal and press **Ctrl + C** to force it to stop!

---

### **🔬 Advanced: The `break` Statement**

Sometimes you want to exit a loop early if an "Emergency" or specific event occurs. We use the `break` keyword for this.

```python
oxygen_level = 100

while oxygen_level > 0:
    oxygen_level -= 10
    print(f"Oxygen at {oxygen_level}%")
    
    if oxygen_level < 20:
        print("⚠️ CRITICAL ALERT: Oxygen too low! Shutting down.")
        break  # Exit the loop immediately

```

---

> **🔬 Research Tip:** `while` loops are essential for **Stochastic Simulations** or **Monte Carlo models** in biology, where you run an experiment until a certain statistical confidence level is reached, rather than for a set number of steps.


## <span style="color:#2774AE;">🔍 The "Glass Box" Method: Debugging with Print</span>

Computers are often "Black Boxes"—we give them input, and they give us output, but we don't see the gears turning in the middle. **Debugging** is the art of turning that black box into a **Glass Box** so we can see exactly what the computer is thinking at every step.

The simplest and most effective tool for this is the **Strategic Print Statement**.

---

### **1. The "Trace" Technique**

When a loop or an `if` statement isn't behaving, place print statements to "trace" the values of variables as they change.

```python
# Problem: This loop is supposed to find the average, but the result is wrong.
data = [10, 20, 30]
total = 0

for value in data:
    # DEBUG PRINT: Check the state of variables inside the loop
    print(f"DEBUG: Current value is {value}, Total before adding was {total}")
    
    total += value
    
    print(f"DEBUG: Total after adding is now {total}")

print(f"Final Average: {total / len(data)}")

```

---

### **2. The "Branch" Check**

If your `if-else` logic isn't working, use prints to see which "branch" the computer is actually taking.

```python
sugar_level = 110

if sugar_level > 120:
    print("Branch A: High")
elif sugar_level > 100:
    print("Branch B: Normal-High") # Is the computer reaching here?
else:
    print("Branch C: Low")

```

* **Reproducibility:** You use the exact same math every time, reducing human error.
* **Organization:** Instead of one giant "wall of code," your script becomes a collection of tools (e.g., `clean_data()`, `run_analysis()`, `save_results()`).
* **Abstraction:** You don't need to remember the complex formula for DNA melting temperature every time; you just call `calculate_tm(sequence)`.

## ## <span style="color:#2774AE;">📊 Handling Real Data: From Lists to CSVs with Pandas</span>

In research, we don't just keep data inside Python's memory. We need to save it to our hard drive so we can open it in Excel, share it with lab mates, or upload it to a database. To do this efficiently, we use **Pandas**.

### **1. What is Pandas?**

Think of Pandas as **"Excel for Python."** * In standard Python, a dataset is just a bunch of nested lists.

* In Pandas, that data is turned into a **DataFrame**—a smart table with labeled rows and columns that makes filtering and math much faster.

---

### **Step 1: Creating and Saving the Data**

Before we can read a file, we need to create one. Let's simulate a small **Single-Cell RNA-seq** matrix.

```python
import pandas as pd
import random

# 1. Define our experimental parameters
n_cells = 5
gene_names = ["GAPDH", "ACTB", "MT-CO1", "SOX2", "NANOG", "POU5F1", "CANX", "B2M", "MALAT1", "TBP"]

# 2. Create a "List of Lists" (Random RNA counts)
# Each sub-list represents one cell's expression for these 10 genes
raw_data = [[random.randint(0, 500) for _ in range(len(gene_names))] for _ in range(n_cells)]

# 3. Convert our "List of Lists" into a Pandas DataFrame
df_to_save = pd.DataFrame(raw_data, columns=gene_names)

# 4. Write to a CSV file
# index_label="Cell_ID" gives our rows a clear name
df_to_save.to_csv("scrna_counts.csv", index_label="Cell_ID")

print("✅ File 'scrna_counts.csv' has been created and saved to your folder!")

```

---

### **Step 2: Reading and Inspecting the Data**

Now, imagine you just received this file from a sequencing core. Here is how you load it.

```python
# 1. Load the data using read_csv
df = pd.read_csv("scrna_counts.csv", index_col="Cell_ID")

# 2. Look at the "Head" (the first few rows)

```

---

### **Step 3: Quick Analysis (The Power of Pandas)**

Because this is a DataFrame and not just a list, we can perform "vectorized" math (math on a whole column at once).

